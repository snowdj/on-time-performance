### Import Packages

```python
import numpy as np
import pandas as pd

import glob

import datetime as dt
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

%matplotlib inline
pd.options.display.max_columns = None # Display all columns
```

### Import Data

```python
# Create path/pattern for data import
path = '../data/'
pattern = '*.csv'
path+pattern
```

```python
# Create csv_file list of data files
csv_files = glob.glob(path + pattern)
csv_files
```

```python
# Import and concatinate into one DataFrame
df = pd.concat((pd.read_csv(f,low_memory=False) for f in csv_files))
```

```python
# Reverse dataframe
df = df.iloc[::-1]
# Reset index due to DataFrame reversal
df = df.reset_index(drop=True)
# Save DataFrame as csv
df.to_csv('../data/df.csv')
```

### Data Cleaning

```python
# Drop 'Unnamed: 17'
df.drop('Unnamed: 17', axis=1, inplace=True)
```

```python
# Change date column to datetime format
df['FL_DATE'] = pd.to_datetime(df['FL_DATE'])
df.info()
```

```python
# Convert int64 columns to object/string
convert_list_int = ['YEAR','MONTH', 'DAY_OF_MONTH']
df[convert_list_int] = df[convert_list_int].astype('object')
```

```python
# Rename the carrier codes to carrier name:
print(df['CARRIER'].value_counts())
carrier_dict = {'WN': 'Southwest',
                'DL': 'Delta',
                'AA': 'American',
                'OO': 'SkyWest',
                'UA': 'United',
                'B6': 'JetBlue',
                'EV': 'ExpressJet',
                'AS': 'Alaska',
                'NK': 'Spirit',
                'F9': 'Frontier',
                'HA': 'Hawaiian',
                'VX': 'Virgin America'}
df['CARRIER'] = df['CARRIER'].replace(carrier_dict)
```

```python
print(df['DAY_OF_WEEK'].value_counts())
weekday_dict = {1: 'Monday',
                2: 'Tuesday',
                3: 'Wednesday',
                4: 'Thursday',
                5: 'Friday',
                6: 'Saturday',
                7: 'Sunday'}
df['DAY_OF_WEEK'] = df['DAY_OF_WEEK'].replace(weekday_dict)
```

```python
# Create list of numeric columns
numeric_columns = df.select_dtypes(exclude=['object','datetime64']).columns
numeric_columns
```

```python
# Create list of categorical columns
cat_columns = df.select_dtypes(include=['object']).columns
cat_columns
```

### EDA Visualization

```python
corr = df.corr()
sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values,cmap='magma')
plt.savefig('../assets/png/02-correlations.png')
```

```python
def draw_histograms(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(10,20))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(ax=ax, bins=50)
        plt.axvline(df[var_name].mean(), color='r', linestyle='dashed', linewidth=2)
        plt.axvline(df[var_name].median(), color='y', linestyle='dashed', linewidth=2)
        ax.set_title(var_name+" Distribution")
    fig.tight_layout()
    plt.savefig('../assets/png/03-histograms.png')
    plt.show()
```

```python
draw_histograms(df, numeric_columns, int(len(numeric_columns)/2)+1, 2)
```

```python
def draw_cat_countplots(df,columns,n_rows,n_cols):
    fig=plt.figure(figsize=(10,20), dpi=80)
    for i, col in enumerate(cat_columns):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        sns.countplot(x=col, data=df, order=df[col].value_counts().index)
        plt.xticks(rotation=90)
        ax.set_title('DEPARTURE DELAYS')
        plt.ylabel('No. of Occurences')
    plt.title('Departure Delays')
    fig.tight_layout()
    plt.savefig('../assets/png/04-cat-countplots.png')
    plt.show()     
```

```python
draw_cat_countplots(df, cat_columns, int(len(cat_columns)/2)+1, 2)
```

```python
# Draw bar plots of categorical data departures
def draw_bars_dep(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(10,20))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        sns.barplot(x=var_name,y='DEP_DELAY',data=df,order=df[var_name].value_counts().index)
        plt.xticks(rotation=90)
        ax.set_title('DEPARTURE DELAYS')
    fig.tight_layout()
    plt.savefig('../assets/png/05-bars-departures.png')
    plt.show()
```

```python
draw_bars_dep(df, cat_columns, int(len(cat_columns)/2)+1, 2)
```

```python
# Draw bar plots of categorical data arrivals
def draw_bars_arr(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(10,20))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        sns.barplot(x=var_name,y='ARR_DELAY',data=df,order=df[var_name].value_counts().index)
        plt.xticks(rotation=90)
        ax.set_title('ARRIVAL DELAYS')
    fig.tight_layout()
    plt.savefig('../assets/06-bars-arrivals.png')
    plt.show()
```

```python
draw_bars_arr(df, cat_columns, int(len(cat_columns)/2)+1, 2)
```

```python
# Draw scatter plots of numerical columns departures
def draw_scatters_dep(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(10,20))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        sns.regplot(x=var_name,y='DEP_DELAY',data=df,fit_reg=False,scatter_kws={'alpha':0.1})
        ax.set_title(var_name +" vs. DEP_DELAY")
    fig.tight_layout()
    plt.savefig('../assets/png/07-scatters-departures.png')
    plt.show()
```

```python
draw_scatters_dep(df, numeric_columns[:-1], int(len(numeric_columns[:-1])/2)+1, 2)
```

```python
# Draw scatter plots of numerical columns arrivals
def draw_scatters_arr(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(10,20))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        sns.regplot(x=var_name,y='ARR_DELAY',data=df,fit_reg=False,scatter_kws={'alpha':0.1})
        ax.set_title(var_name +" vs. ARR_DELAY")
    fig.tight_layout()
    plt.savefig('../assets/png/08-scatters-arrivals.png')
    plt.show()
```

```python
draw_scatters_arr(df, numeric_columns[:-1], int(len(numeric_columns[:-1])/2)+1, 2)
```
